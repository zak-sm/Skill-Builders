```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler # Used to standardize the data to a common range
from sklearn.model_selection import train_test_split
from sklearn import svm 
from sklearn.metrics import accuracy_score



```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df. describe()
```

```python
df['Outcome'].value_counts()
```

0 = Non-Diabetic 1=Diabetic

```python
df.groupby('Outcome').mean()
```

```python
fac_x = df.drop(columns=['Outcome'])
fac_y = df['Outcome']
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(fac_x)
```

```python
standardized_data = scaler.transform(fac_x)
```

```python
X = standardized_data
```

Training the Model

```python
train_x, test_x, train_y, test_y = train_test_split(X, fac_y, test_size=0.2, stratify=fac_y, random_state=2)
```

```python
train_y.shape
```

```python
classifier = svm.SVC(kernel='linear')
```

```python
# Training the support vector classifier
classifier.fit(train_x, train_y)
```

Model Evaluation

Accuracy Score 

```python
# ccuracy score on the training data
x_train_predict = classifier.predict(train_x)
training_data_accuracy = accuracy_score(x_train_predict, train_y)
```

```python
print (training_data_accuracy)
```

Using the model to predict the testing data


```python
X_test_predict = classifier.predict(test_x)
test_data_accuracy = accuracy_score(X_test_predict, test_y)
print (test_data_accuracy)
```

Making a Predictive System

```python
input_data = (8,99,84,0,0,35.4,0.388,50)

# change input to a numpy array
input_data_np = np.asarray(input_data)

# reshape the array as per the requirement of the model | We are only predicting one point
input_data_reshape = input_data_np.reshape(1, -1)

# standardize the input data
input_data_std = scaler.transform(input_data_reshape)

# predict the output
prediction = classifier.predict(input_data_std)

if prediction[0] == 0:
    print('The person is not diabetic')
else:
    print('The person is diabetic')
```
